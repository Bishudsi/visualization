```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)
z = np.random.randint(0,100,50)
fig, ax =plt.subplots(figsize=(5,3))
#ax.bar(x,y)
#ax.bar(x,y)
ax.hist(y)
#ax.plot(x,y)
```

```python
font1 = {'family':'sans-serif','color':'blue','size':18,'weight':'bold'}
font2 = {'family':'monospace','color':'green','size':14}

fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y,
           marker = '*',
           color = 'indigo')
```

```python
fig,ax = plt.subplots(figsize=(5,3))

ax.plot(x,
           y,
           marker = '*',
           color = '#7425B9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = "#B93059",
           markerfacecolor = '#000000')

ax.grid(axis = 'y', color = "#73D934",linewidth = 3, linestyle = '--')
```

```python
fig,ax = plt.subplots(figsize=(8,6))

ax.scatter(x,
           y,
           z,
           cmap="Blues",
           alpha=0.4,
           edgecolors="grey",
           linewidth=4)

ax.set_xlabel("the X axis")
ax.set_ylabel("the Y axis")
ax.set_title("Distant galaxies from James Webb")
```
